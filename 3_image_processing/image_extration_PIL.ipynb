### Solve with PIL Image

```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
root_path = os.getcwd()
background_path = os.path.join(root_path, "test_imgs", "background.jpg" )
foreground_path = os.path.join(root_path, "test_imgs", "foreground.jpg")
mask_path = os.path.join(root_path, "test_imgs", "mask.png")
```

```python
pil_background = Image.open(background_path)
pil_foreground = Image.open(foreground_path)
pil_mask = Image.open(mask_path)

if type(pil_background) is None or type(pil_foreground) is None or type(pil_mask) is None:
    print("... Invalid Image paths ...")
else:
    print("Success")
```

```python
width, height = pil_foreground.size
print(width, height)


pil_background_copy = pil_background.copy()

pil_mask_pixel = pil_mask.load()
pil_foreground_pixel = pil_foreground.load()
pil_background_pixel = pil_background_copy.load()


for row in range(width):
    for col in range(height):
        if pil_mask_pixel[row, col] == (255,0,0):
            pil_background_pixel[row, col] = pil_foreground_pixel[row, col]

np_image = np.array(pil_background_copy)
plt.imshow(np_image)
```
