### Solve with OpenCV

```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
root_path = os.getcwd()
background_path = os.path.join(root_path, "test_imgs", "background.jpg")
foreground_path = os.path.join(root_path, "test_imgs", "foreground.jpg")
mask_path = os.path.join(root_path, "test_imgs", "mask.png")

cv_background = cv2.imread(background_path)
cv_foreground = cv2.imread(foreground_path)
cv_mask = cv2.imread(mask_path)

if cv_background is None or cv_foreground is None or cv_mask is None:
    print("...Invalid Image paths...")
else:
    print("Success")
```

```python
print(cv_mask.shape)

# BGR
# Blue
green_channel = cv_mask[:,:,0]
# Green
blue_channel = cv_mask[:,:,1]
# Red
red_channel = cv_mask[:,:,2]
plt.imshow(red_channel)

width, height = red_channel.shape

```

```python
count = 0

cv_background_copy = cv_background.copy()


for row in range(width):
    for col in range(height):
        cur_red_value = red_channel[row, col]
        if cur_red_value == 255:
            count = count + 1 
            cv_background_copy[row, col] = cv_foreground[row, col]

print(count)
plt.imshow(cv_background_copy)
```
